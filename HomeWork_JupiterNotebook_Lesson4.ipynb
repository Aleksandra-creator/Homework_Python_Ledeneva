*Выводы:*
    
*- Стоимость домов значительно зависит от жилой площади, где имеет пропорциональную зависимость: чем больше количество квадрвтных метров, тем выше стоимость дома. Однако наблюдаются исключения, которые зависят от других показателей*

*- Стоимость дома так же сильно зависит от количества ванных комнат, самые дорогие дома имеют 2-3 ванных комнаты*

*- Стоимость зависит от количества этажей дома, при этом самый дорогой дом имеет 2,5 этажа из 3,5 возможных*

*- Стоимость зависит от оценки (grade) дома, здесь наблюдаем прямую зависимость: чем выше grade, тем дороже стоимость дома. Однако, на графике наблюдается небольшое количество выбросов*

*- Среднюю зависимость на стоимость дома имеют показатели: количества спальных комнат, наличие и площадь подвальных помещений, долгота по расположению дома, а также вид*

*- Низкую зависимость на стоимость дома имеют такие показатели, как: индекс, год реновации и широта по расположению дома*
